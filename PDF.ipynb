```python
from bs4 import BeautifulSoup
```

```python
with open('pg_0001.html', 'r') as f:
    contents = f.read()
    #,encoding="utf-8"
```

```python
soup = BeautifulSoup(contents, "html.parser")
```

```python
soup
```

```python
TAG_SPAN = soup.find('body')
#TAG_SPAN

FIND_THE_SPECIFIC_TAG = TAG_SPAN.findAll('span', style='font-family: Calibri-Bold; font-size:11px') 
```

```python

data = []
for j in FIND_THE_SPECIFIC_TAG:
    #print(j.text)
    data.append(j.text)
#data

DF_1 = ''.join(data)
DF_1
```
